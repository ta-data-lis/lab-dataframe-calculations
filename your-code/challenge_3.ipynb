## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub-problem-1
# Finding the total sum of amount_spent by each customer id
df_cust_spent = orders.groupby('CustomerID').amount_spent.sum()
df_cust_spent
```

```python
# Sub-problem-2
q1 = df_cust_spent.quantile(0.25) # First quantile
q3 = df_cust_spent.quantile(0.75) # Third quantile
iqr = q3-q1 #Inter-quantile Range

lower_bound = q1 -(1.5 * iqr) 
upper_bound = q3 +(1.5 * iqr)

# Selecting customers whose aggregated amount_spent is in a given quantile range
reqd_customers = []
for cust_id in df_cust_spent.index:
  if df_cust_spent[cust_id] > lower_bound and df_cust_spent[cust_id] < upper_bound:
    reqd_customers.append(cust_id)

# Printing customers whose aggregated amount_spent is in a given quantile range
print(reqd_customers)
```

```python
# Sub-problem-3
vip_level = df_cust_spent.quantile(0.95) # VIP quantile
pref_level = df_cust_spent.quantile(0.75) # Preferred quantile

# Selecting VIP and Preferred Customers
vip_customers = []
pref_customers = []

for cust_id in df_cust_spent.index:
  # For VIP level
  if df_cust_spent[cust_id] > vip_level:
    vip_customers.append(cust_id)
  # For Preferred level
  elif df_cust_spent[cust_id] < vip_level and df_cust_spent[cust_id] > pref_level:
    pref_customers.append(cust_id)

# Printing VIP and preferred customers
print("Vip customers are:-> ", vip_customers)
print("Preferred customers are:-> ", pref_customers)
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
VIP_country = {}

for cust in vip_customers:
  temp = orders[orders.CustomerID == cust].Country.values[0]
  VIP_country[temp] = VIP_country.get(temp, 0) + 1

VIP_country
```

```python
# country has the most VIP Customers
print("country has the most VIP Customers: ", list(dict(sorted(VIP_country.items(), key=lambda item: item[1])).keys())[-1])
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
combined_country = {}
combined_customers = vip_customers + pref_customers

for cust in combined_customers:
  temp = orders[orders.CustomerID == cust].Country.values[0]
  combined_country[temp] = combined_country.get(temp, 0) + 1

combined_country
```

```python
# country has the most VIP+Preferred Customers combined
print("country has the most VIP+Preferred Customers combined: ", list(dict(sorted(combined_country.items(), key=lambda item: item[1])).keys())[-1])
```
